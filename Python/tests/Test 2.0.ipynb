```python
#LOOPS


# Universal Input Lists
numbers = [5, -2, 12, 7, 0, 33, -10, 8]
words = ["apple", "banana", "kiwi", "fig", "strawberry", "grape"]

# 1. Function to divide two numbers with divide-by-zero handling
def divide(a, b):
    try:
        result = a / b
        print(f"{a} / {b} = {result}")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

divide(numbers[0], numbers[2])  
divide(numbers[0], numbers[4])  

print("\n" + "-"*50)

# 2. Ask user for a number, handle non-number input
def ask_number():
    try:
        user_input = input("Enter a number: ")
        number = float(user_input)
        print("You entered:", number)
    except ValueError:
        print("Error: That is not a valid number.")

print("\n" + "-"*50)

# 3. Open a file using try, print custom error if file not found
def open_file():
    try:
        with open("sample.txt", "r") as f:
            print("File contents:\n", f.read())
    except FileNotFoundError:
        print("Error: File 'sample.txt' was not found.")

open_file()

print("\n" + "-"*50)

# 4. Function that adds two numbers, raises error if input is not numeric
def safe_add(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both inputs must be numbers.")
    return a + b

try:
    result = safe_add(numbers[1], numbers[3])  
    print("Sum:", result)
except TypeError as e:
    print("Error:", e)

try:
    result = safe_add(numbers[1], words[0])  
    print("Sum:", result)
except TypeError as e:
    print("Error:", e)

print("\n" + "-"*50)

# 5. Safely read an integer from user
def read_integer():
    try:
        val = int(input("Enter an integer: "))
        print("You entered:", val)
    except ValueError:
        print("Error: That is not a valid integer.")

```

```python
#LAMMDA, MAP AND FILTER

numbers = [5, -2, 12, 7, 0, 33, -10, 8]

# 1. Lambda to add 10
add_10 = lambda x: x + 10
print("5 + 10 =", add_10(5)) 

# 2. Map to square numbers
squares = list(map(lambda x: x**2, numbers))
print("Squares:", squares)  

# 3. Filter even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", evens)  

# 4. Lowercase to uppercase
words = ["apple", "banana", "kiwi", "fig", "strawberry", "grape"]
upper_words = list(map(lambda x: x.upper(), words))
print("Uppercase words:", upper_words)  

# 5. Filter > 50
greater_than_50 = list(filter(lambda x: x > 50, numbers))
print("Numbers > 50:", greater_than_50)  
```

```python
#VARIABLE SCOPE


# 1. Use of a global variable inside a function
print("1: Global Variable Inside Function")

x = 100  # global variable

def use_global():
    global x
    x += 20
    print("Inside function (x):", x)

use_global()
print("Outside function (x):", x)
print("\n" + "-"*50)

# 2. Local variable does not affect global variable
print("2: Local Does Not Affect Global")

y = 200  

def local_different():
    y = 50  
    print("Inside function (y):", y)

local_different()
print("Outside function (y):", y)
print("\n" + "-"*50)

# 3. Create a function that modifies a global variable
print("3: Modify Global Variable")

count = 0

def increment():
    global count
    count += 1
    print("Inside function (count):", count)

increment()
increment()
print("Outside function (count):", count)
print("\n" + "-"*50)

# 4. One function sets local, one accesses global
print("4: Local vs Global Functions")

name = "GlobalName"

def set_local():
    name = "LocalName"
    print("Inside set_local (name):", name)

def access_global():
    print("Inside access_global (name):", name)

set_local()
access_global()
print("\n" + "-"*50)

# 5. Variable declared inside a function is not accessible outside
print(" 5: Local Variable Not Accessible Outside")

def create_var():
    secret = "This is inside"
    print("Inside function (secret):", secret)

create_var()

try:
    print("Outside function (secret):", secret)
except NameError as e:
    print("Error:", e)
```
